### Mengakses Baris (iloc)

```python
# Untuk mengakses baris kamu dapat menggunakan syntax berikut 
# df.iloc[baris_ke]

df.iloc[0]
```

```python
# Untuk mengakses range tertentu
df.iloc[:3]
```

```python
# Untuk mengakses kolom dan range tertentu

# df.iloc[start:end, list kolom apa aja]
df.iloc[1:5,[0,3,4]]
```

### Mengakses Baris dengan nama kolomya (loc)

```python
# Syntaxnya begini
# df.loc[start:end, list nama kolom apa aja]

df.loc[1:5,["PassengerId","Name","Sex"]]
```

```python
# Challange 3 :
# Coba akses baris 5:10 dengan menampilkan kolom ke 1 dengan terakhir saja
```

### Filtering

Untuk melakukan filter, kita dapat menggunakan beberapa cara

```python
# Dengan membuat indexing
# df[(kondisi 1) (kondisi 2)]

# Contoh
df[df.Age > 5]
```

```python
# Untuk mengakses baris tertentu kita dapat menggunakan loc 

# df.loc[kondisi,list kolom]
df.loc[df.Age>50,["Survived"]]
```

```python
# Kita juga dapat menggabungkan kondisi dengan menggunkan && (and) atau || (or)
# dengan syntax 
# df[(condition 1) | (conditon 2) | (dsb )]
# df[(condition 1) & (conditon 2) & (dsb )]

df[(df.Age > 10) & (df.Age <50)]
```

```python
# Challange 4:
# Coba tampilkan Cabin yang tidak Farenya  yang dibawah dari 10
```

### Mengubah Data

Terdapat beberapa cara untuk mengubah data diantaranya

```python
# Agar dataset tidak berubah kita dapat mengkopy dulu datasetnya dalam suatu kolom
df2 = df.copy(deep=True) #deep bertujuan agar kita melakukan hard copy terhadap data
```

```python
# Untuk mengubah nilai suatu kolom kamu dapat mendefiniskannya menggunakan tanda =
# df[kolom] = pengganti kolom

print("Ini adalah data lama :",df2.Age)

print("Bila Kita Ubah ")

df2.Age = 3033

df2.Age
```

Terlihat diatas bahwa kita sudah mengubah dataset yang ada

```python
# Kita juga dapat menggunakan rata-rata ,median, dsb sebagai replacement
df2.Age = df.Age # Untuk mengembalikan data terlebih dahulu

df2.Age = df2.Age.mean() # Untuk replace ke rata-rata
```

```python
print(df2.Age)
```

### Fill Missing Value

```python
# Dalam dataset terkadang terdapat missing value yang harus diisi terlebih dahulu

df2.isna()
# Apabila datasetnya menunjukan True maka itu adalah missing value
```

```python
# Untuk mereplacenya kita dapat menggunakan fillna

df.Fare = df.Fare.fillna(df.Fare.mean())
```

# Seaborn

```python
# Seaborn memiliki pola yang cukup mudah dipahami dalam visualisasi data
# Untuk mengimport seaborn kamu dapat melakukan hal ini
import seaborn as sns
```

```python
# Kemudian untuk menampilkan data, kamu tinggal melakukan syntax seperti berikut
"""
    sns.JENIS_PLOT(data=SUMBER_DATA_FRAME, x=X_AXIS,Y=y_axis)
"""

# contoh 
sns.boxplot(data=df,x="Fare")
```

```python
# Kamu juga dapat melakukan pemisahan dengan menggunkana hue
sns.scatterplot(data=df,x="Age",y="Fare",hue="Survived")
```
